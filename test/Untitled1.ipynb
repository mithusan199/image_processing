```python
from skimage import io, img_as_ubyte
from scipy import ndimage
import numpy as np
from matplotlib import pyplot as plt
```

```python
print(img.shape)
```

```python
img=img_as_ubyte(io.imread("WhatsApp Image 2023-04-22 at 00.09.03.jpeg",as_gray=False))
```

```python
flippedLR=np.fliplr(img)
```

```python
flippedUD=np.flipud(img)
```

```python
plt.subplot(2,1,1)
plt.imshow(img, cmap="hsv")
plt.subplot(2,2,4)
plt.imshow(flippedUD)
plt.subplot(2,2,3)
plt.imshow(flippedLR)
```

```python
rotated=ndimage.rotate(img,45)
```

```python
plt.imshow(rotated)
```

```python
uniform_fil=ndimage.uniform_filter(img,size=20)
plt.imshow(uniform_fil)
```

```python
gauss_fil=ndimage.gaussian_filter(img,sigma=11)
plt.imshow(gauss_fil)
```

```python
m_fil=ndimage.median_filter(img,2)
plt.imshow(m_fil)
```
